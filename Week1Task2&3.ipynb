```python
df=pd.read_csv('/content/drive/MyDrive/SentimentAnalysisDataset2.csv', on_bad_lines='skip') 
```

```python
df.head()
```

```python
df=df.drop(['ItemID','SentimentSource'],axis=1)
```

```python
import string
string.punctuation
```

```python
def remove_punct(text):
  punctfree="".join([i for i in text if i not in string.punctuation])
  return punctfree

df['Punctuation-free']=df['SentimentText'].apply(lambda x: remove_punct(x))
```

```python
df.head()
```

```python
df['Lower']=df['Punctuation-free'].apply(lambda x: x.lower())
```

```python
df
```

```python
import nltk
```

```python
nltk.download('punkt')
```

```python
df['Tokenized-words']=df['Lower'].apply(lambda x: nltk.word_tokenize(x))
```

```python
df
```

```python
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output
```

```python
df['no_stopwords']= df['Tokenized-words'].apply(lambda x:remove_stopwords(x))
```

```python
df
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
porter_stemmer = PorterStemmer()
```

```python
def stemming(text):
  stem_text = [porter_stemmer.stem(word) for word in text]
  return stem_text
df['msg_stemmed']=df['no_stopwords'].apply(lambda x: stemming(x))
```

```python
df
```

```python
from nltk.stem import WordNetLemmatizer

wordnet_lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
nltk.download('omw-1.4')
```

```python
def lemmatizer(text):
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    return lemm_text
df['msg_lemmatized']=df['no_stopwords'].apply(lambda x:lemmatizer(x))
```

```python
df2=pd.DataFrame(columns=['Tweet','Label'])
df2['Tweet']=df['Lower']
df2['Label']=df['Sentiment']
```

```python
df2
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 

from sklearn.feature_extraction.text import CountVectorizer


cv = CountVectorizer(stop_words = 'english')
words = cv.fit_transform((df2.Tweet))


sum_words = words.sum(axis=0)

words_freq = [(word, sum_words[0, i]) for word, i in cv.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)


wordcloud = WordCloud(background_color = 'white', width = 1000, height = 1000).generate_from_frequencies(dict(words_freq))

plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.title("WordCloud - Vocabulary from Reviews", fontsize = 22)
```

```python
general_words =' '.join([text for text in df2['Tweet'][df2['Label'] == 1]])
wordcloud = WordCloud(width=800, height=500, random_state = 0, max_font_size = 110).generate(general_words)
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.title('General Words')
plt.show()
```

```python
neg_words =' '.join([text for text in df2['Tweet'][df2['Label'] == 0]])
wordcloud = WordCloud(width=800, height=500, random_state = 0, max_font_size = 110).generate(neg_words)
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.title('General Words')
plt.show()
```
